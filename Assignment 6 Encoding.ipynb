```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
zone = ['southwest', 'southeast', 'northwest', 'northeast']
```

```python
ordinal = OrdinalEncoder(categories=[zone])
encoded = ordinal.fit_transform(df7[['region']])
```

```python
encoded
```

```python
new_region = pd.DataFrame(encoded, columns=['Locality'])
```

```python
new_region.head()
```

```python
df7.head()
```

```python
df7.region = new_region
```

```python
df7.head()
```
